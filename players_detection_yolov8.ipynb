```python
from ultralytics import YOLO

# YOLOv8 nano modelini başlat
model = YOLO("yolov8n.pt")

# Eğitimi başlat (dataset.location indirdiğin dosyanın yolunu otomatik bulur)
results = model.train(
    data=f"{dataset.location}/data.yaml",
    epochs=25,       # Eğitim döngüsü sayısı. Gerekirse daha sonra 50 veya 100'e çıkarabilirsin.
    imgsz=640,       # Görüntü boyutu
    batch=16         # Tek seferde işlenecek görsel sayısı
)
```

```python
# 1. Eğittiğin en iyi ağırlıkları (best.pt) yükle
model = YOLO("runs/detect/train/weights/best.pt")

# 2. Videoyu tahmin et ve kaydet
# "mac_videosu.mp4" yerine kendi yüklediğin videonun adını yazmalısın.
# conf=0.25 değeri güven eşiğidir (0 ile 1 arası). Doğruluğu artırmak için bunu 0.40 veya 0.50 yapabilirsin.
results = model.predict(source="/content/deneme.mp4", save=True, conf=0.25)
```
